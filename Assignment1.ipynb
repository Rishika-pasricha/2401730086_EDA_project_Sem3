```python
# ASSIGNMENT-1

import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from scipy import stats
```

```python
# Q1: Bike Prices Analysis


bike_prices = (12000, 15000, 18000, 20000, 50000)

# (a) Calculate Mean
mean_price = np.mean(bike_prices)
print(f"\n(a) Mean Price:")
print(f"    Formula: Sum of all values / Number of values")
print(f"    Calculation: {sum(bike_prices)} / {len(bike_prices)}")
print(f"    Mean = ₹{mean_price:,.2f}")

# (b) Calculate Median
median_price = np.median(bike_prices)
print(f"\n(b) Median Price:")
print(f"    Sorted data: {sorted(bike_prices)}")
print(f"    Middle value (position 3): ₹{median_price:,.2f}")

# (c) Which average to advertise
print(f"\n(c) Which 'Average' to Advertise?")
print(f"    Mean = ₹{mean_price:,.2f}")
print(f"    Median = ₹{median_price:,.2f}")
print(f"    RECOMMENDATION: Advertise MEDIAN (₹{median_price:,.2f})")
print(f"    WHY? It's lower and makes bikes seem more affordable.")
print(f"    The mean is inflated by the expensive bike (₹50,000).")

```

```python
# Q2: Left-Skewed Distribution
print("\n" + "=" * 70)
print("Q2: LEFT-SKEWED DISTRIBUTION")
print("=" * 70)

print("\nIn a LEFT-SKEWED distribution:")
print("  • Shape: Long tail extends to the LEFT")
print("  • Relationship: Mode > Median > Mean")
print("  • The mean is pulled LEFT by extreme low values")
print("  • Most data clusters on the RIGHT (high values)")
print("\nExample: If most students scored 80-100, but a few scored 20-30,")
print(" the distribution would be left-skewed.")
```

```python
# ============================================================================
# Q3: Exam Scores Analysis
# ============================================================================
print("\n" + "=" * 70)
print("Q3: EXAM SCORES ANALYSIS")
print("=" * 70)

exam_scores = [85, 92, 78, 90, 85, 67, 88, 95, 85, 72]
sorted_scores = sorted(exam_scores)

# (a) Calculate Mean, Median, Mode
mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)
mode_result = stats.mode(exam_scores, keepdims=True)
mode_score = mode_result.mode[0]

print(f"\n(a) Measures of Central Tendency:")
print(f"    Data: {exam_scores}")
print(f"    Sorted: {sorted_scores}")
print(f"\n    Mean = {mean_score:.2f}")
print(f"    Median = {median_score:.2f}")
print(f"    Mode = {mode_score} (appears {Counter(exam_scores)[mode_score]} times)")

# (b) Shape of distribution
print(f"\n(b) Shape of Distribution:")
print(f"    Mean ({mean_score:.2f}) < Median ({median_score:.2f})")
print(f"    CONCLUSION: LEFT-SKEWED distribution")
print(f"    JUSTIFICATION: Mean is pulled down by low outliers (67, 72)")

# (c) Five-Number Summary
q1 = np.percentile(sorted_scores, 25)
q2 = np.percentile(sorted_scores, 50)  # median
q3 = np.percentile(sorted_scores, 75)
minimum = min(sorted_scores)
maximum = max(sorted_scores)

print(f"\n(c) Five-Number Summary:")
print(f"    Minimum: {minimum}")
print(f"    Q1 (25th percentile): {q1}")
print(f"    Q2 (Median, 50th percentile): {q2}")
print(f"    Q3 (75th percentile): {q3}")
print(f"    Maximum: {maximum}")
print(f"\n    IQR (Interquartile Range) = Q3 - Q1 = {q3 - q1}")
```

```python
# ============================================================================
# Q4: Population Parameter vs Sample Statistic
# ============================================================================
print("\n" + "=" * 70)
print("Q4: POPULATION PARAMETER vs SAMPLE STATISTIC")
print("=" * 70)

print("\nPOPULATION PARAMETER:")
print("  • Describes the ENTIRE population")
print("  • Usually unknown (can't measure everyone)")
print("  • Denoted by Greek letters: μ (mu), σ (sigma)")
print("  • Example: Average height of ALL students in India")

print("\nSAMPLE STATISTIC:")
print("  • Describes a SAMPLE from the population")
print("  • Used to estimate population parameters")
print("  • Denoted by Roman letters: x̄ (x-bar), s")
print("  • Example: Average height of 100 randomly selected students")

# Example demonstration
population = np.random.normal(170, 10, 10000)  # 10,000 students
sample = np.random.choice(population, 100, replace=False)  # 100 students

print(f"\nDEMONSTRATION:")
print(f"  Population mean (μ): {np.mean(population):.2f} cm")
print(f"  Sample mean (x̄): {np.mean(sample):.2f} cm")
print(f"  Sample estimates the population!")
```

```python
# ============================================================================
# Q5: Employee Salaries Analysis
# ============================================================================
print("\n" + "=" * 70)
print("Q5: EMPLOYEE SALARIES ANALYSIS")
print("=" * 70)

salaries = [45000, 50000, 55000, 60000, 250000]

# (a) Calculate Mean and Median
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)

print(f"\n(a) Mean and Median Salary:")
print(f"    Salaries: {salaries}")
print(f"    Mean = ${mean_salary:,.2f}")
print(f"    Median = ${median_salary:,.2f}")

# (b) Which measure to use
print(f"\n(b) Which Measure to Use?")
print(f"\n    COMPANY would advertise: MEAN (${mean_salary:,.2f})")
print(f"      Why? Sounds more impressive and attractive")
print(f"\n    EMPLOYEE would prefer: MEDIAN (${median_salary:,.2f})")
print(f"      Why? Shows the 'typical' salary accurately")
print(f"      The mean is inflated by one outlier (${salaries[-1]:,} - CEO?)")
print(f"      4 out of 5 employees earn ${salaries[-2]:,} or less!")
```

```python
# ============================================================================
# Q7: Graphical Representation - Learning Modes
# ============================================================================
print("\n" + "=" * 70)
print("Q7: GRAPHICAL REPRESENTATION - LEARNING MODES")
print("=" * 70)

raw_data = ['F', 'H', 'O', 'F', 'F', 'O', 'H', 'F', 'O', 'F', 
            'H', 'F', 'F', 'F', 'O', 'H', 'O', 'F', 'H', 'F']

# (a) Create Frequency Table
freq_count = Counter(raw_data)
total_students = len(raw_data)

print(f"\n(a) Frequency Table:")
print(f"    {'Mode':<20} {'Frequency':<12} {'Percentage'}")
print(f"    {'-'*50}")
for mode in ['F', 'H', 'O']:
    count = freq_count[mode]
    percentage = (count / total_students) * 100
    mode_name = {'F': 'Face-to-Face', 'H': 'Hybrid', 'O': 'Online'}[mode]
    print(f"    {mode_name:<20} {count:<12} {percentage:.1f}%")
print(f"    {'-'*50}")
print(f"    {'Total':<20} {total_students:<12} 100.0%")

# (b) Create Bar Chart
modes = ['Face-to-Face', 'Hybrid', 'Online']
frequencies = [freq_count['F'], freq_count['H'], freq_count['O']]
colors = ['#3498db', '#2ecc71', '#e74c3c']

plt.figure(figsize=(10, 6))
bars = plt.bar(modes, frequencies, color=colors, edgecolor='black', linewidth=1.5)
plt.xlabel('Learning Mode', fontsize=12, fontweight='bold')
plt.ylabel('Number of Students', fontsize=12, fontweight='bold')
plt.title('Student Preferred Mode of Learning (N=20)', fontsize=14, fontweight='bold')
plt.ylim(0, 12)
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.savefig('learning_modes_bar_chart.png', dpi=300, bbox_inches='tight')
print(f"\n(b) Bar Chart created and saved as 'learning_modes_bar_chart.png'")

# (c) Most popular mode
most_popular = max(freq_count, key=freq_count.get)
most_popular_name = {'F': 'Face-to-Face', 'H': 'Hybrid', 'O': 'Online'}[most_popular]
most_popular_count = freq_count[most_popular]
percentage = (most_popular_count / total_students) * 100

print(f"\n(c) Most Popular Learning Mode:")
print(f"    {most_popular_name} with {most_popular_count} students ({percentage:.1f}%)")

# (d) Why not histogram
print(f"\n(d) Why NOT a Histogram?")
print(f"    • Histograms are for CONTINUOUS NUMERICAL data")
print(f"    • This data is CATEGORICAL (F, H, O are labels, not numbers)")
print(f"    • Histograms have TOUCHING bars (show continuity)")
print(f"    • Bar charts have SEPARATED bars (show distinct categories)")
print(f"    • No meaningful numerical order exists for these categories")
```
